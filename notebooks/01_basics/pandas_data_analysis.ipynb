```python
import pandas as pd
import numpy as np

print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

## 1ï¸âƒ£ GroupBy Operations

Group data by categories and compute statistics.

```python
# Create sample sales dataset
sales = pd.DataFrame({
    'Region': ['East', 'West', 'East', 'North', 'South', 'West', 'East', 'North'],
    'Product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
    'Sales': [100, 150, 120, 200, 180, 110, 90, 160],
    'Quantity': [10, 15, 12, 20, 18, 11, 9, 16],
    'Profit': [20, 30, 25, 40, 35, 22, 18, 32]
})

print("Sales Dataset:")
print(sales)
```

```python
# Group by single column
region_groups = sales.groupby('Region')

print("Total Sales by Region:")
print(region_groups['Sales'].sum())

print("\nAverage Sales by Region:")
print(region_groups['Sales'].mean())

print("\nCount of transactions by Region:")
print(region_groups.size())
```

```python
# Multiple aggregations
print("Multiple aggregations by Region:")
print(region_groups['Sales'].agg(['sum', 'mean', 'max', 'min', 'count']))
```

```python
# Group by multiple columns
region_product = sales.groupby(['Region', 'Product'])

print("Sales by Region and Product:")
print(region_product['Sales'].sum())

print("\nAs DataFrame:")
print(region_product['Sales'].sum().reset_index())
```

```python
# Different aggregations for different columns
print("Custom aggregations:")
result = sales.groupby('Region').agg({
    'Sales': ['sum', 'mean'],
    'Quantity': 'sum',
    'Profit': ['sum', 'max']
})
print(result)
```

### ðŸŽ¯ ML Example: Feature Aggregation

```python
# Customer transaction data
transactions = pd.DataFrame({
    'customer_id': [1, 1, 1, 2, 2, 3, 3, 3, 3],
    'amount': [100, 200, 150, 50, 80, 300, 120, 90, 200],
    'category': ['A', 'B', 'A', 'C', 'A', 'B', 'C', 'A', 'B']
})

print("Transaction data:")
print(transactions)

# Create customer-level features
customer_features = transactions.groupby('customer_id').agg({
    'amount': ['sum', 'mean', 'std', 'count'],
    'category': lambda x: x.nunique()  # Number of unique categories
})

customer_features.columns = ['total_spent', 'avg_transaction', 'std_transaction', 
                              'num_transactions', 'num_categories']
customer_features = customer_features.reset_index()

print("\nCustomer features for ML:")
print(customer_features)
```

## 2ï¸âƒ£ Pivot Tables

```python
# Create pivot table
pivot = pd.pivot_table(sales, 
                       values='Sales', 
                       index='Region', 
                       columns='Product',
                       aggfunc='sum',
                       fill_value=0)

print("Pivot Table (Sales by Region and Product):")
print(pivot)

# Add margins (totals)
pivot_margins = pd.pivot_table(sales, 
                               values='Sales', 
                               index='Region', 
                               columns='Product',
                               aggfunc='sum',
                               fill_value=0,
                               margins=True,
                               margins_name='Total')

print("\nWith totals:")
print(pivot_margins)
```

```python
# Multiple aggregation functions
pivot_multi = pd.pivot_table(sales,
                             values='Sales',
                             index='Region',
                             columns='Product',
                             aggfunc=['sum', 'mean'],
                             fill_value=0)

print("Multiple aggregations:")
print(pivot_multi)
```

## 3ï¸âƒ£ Merging and Joining DataFrames

```python
# Create two related datasets
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [25, 30, 35, 28]
})

orders = pd.DataFrame({
    'order_id': [101, 102, 103, 104, 105],
    'customer_id': [1, 2, 1, 3, 2],
    'amount': [100, 200, 150, 300, 250]
})

print("Customers:")
print(customers)
print("\nOrders:")
print(orders)
```

```python
# Inner join (only matching rows)
inner = pd.merge(customers, orders, on='customer_id', how='inner')
print("Inner Join:")
print(inner)
```

```python
# Left join (all customers, even without orders)
left = pd.merge(customers, orders, on='customer_id', how='left')
print("Left Join:")
print(left)
```

```python
# Right join (all orders, even if customer info missing)
right = pd.merge(customers, orders, on='customer_id', how='right')
print("Right Join:")
print(right)
```

```python
# Outer join (all records from both)
outer = pd.merge(customers, orders, on='customer_id', how='outer')
print("Outer Join:")
print(outer)
```

### ðŸŽ¯ ML Example: Feature Joining

```python
# User features
user_features = pd.DataFrame({
    'user_id': [1, 2, 3, 4, 5],
    'age': [25, 30, 35, 28, 32],
    'country': ['US', 'UK', 'US', 'CA', 'UK']
})

# User behavior
user_behavior = pd.DataFrame({
    'user_id': [1, 1, 2, 3, 3, 4],
    'clicks': [10, 15, 20, 5, 8, 12]
})

# Aggregate behavior
behavior_agg = user_behavior.groupby('user_id').agg({
    'clicks': ['sum', 'mean', 'count']
})
behavior_agg.columns = ['total_clicks', 'avg_clicks', 'sessions']
behavior_agg = behavior_agg.reset_index()

# Merge features
ml_features = pd.merge(user_features, behavior_agg, on='user_id', how='left')
ml_features = ml_features.fillna(0)

print("ML Features Dataset:")
print(ml_features)
```

## 4ï¸âƒ£ Apply and Transform

```python
# Apply custom function to column
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'score': [85, 92, 78]
})

# Apply function
def grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    else:
        return 'C'

df['grade'] = df['score'].apply(grade)
print("With grades:")
print(df)

# Lambda function
df['passed'] = df['score'].apply(lambda x: 'Yes' if x >= 80 else 'No')
print("\nWith pass/fail:")
print(df)
```

```python
# Apply to multiple columns
df_multi = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

# Row-wise apply
df_multi['sum'] = df_multi.apply(lambda row: row['A'] + row['B'] + row['C'], axis=1)
df_multi['max'] = df_multi.apply(lambda row: max(row['A'], row['B'], row['C']), axis=1)

print("Row-wise operations:")
print(df_multi)
```

## ðŸŽ¯ Practice Exercise: Complete Analysis Pipeline

```python
# E-commerce dataset
np.random.seed(42)
ecommerce = pd.DataFrame({
    'order_id': range(1, 101),
    'customer_id': np.random.randint(1, 21, 100),
    'product_category': np.random.choice(['Electronics', 'Clothing', 'Books', 'Home'], 100),
    'amount': np.random.randint(10, 500, 100),
    'quantity': np.random.randint(1, 5, 100)
})

print("E-commerce orders (first 10):")
print(ecommerce.head(10))

# Analysis tasks:

# 1. Total revenue by category
revenue_by_category = ecommerce.groupby('product_category')['amount'].sum().sort_values(ascending=False)
print("\n1. Revenue by Category:")
print(revenue_by_category)

# 2. Customer spending summary
customer_summary = ecommerce.groupby('customer_id').agg({
    'amount': ['sum', 'mean', 'count'],
    'quantity': 'sum'
})
customer_summary.columns = ['total_spent', 'avg_order', 'num_orders', 'total_items']
customer_summary = customer_summary.reset_index()

print("\n2. Top 5 Customers by Total Spent:")
print(customer_summary.nlargest(5, 'total_spent'))

# 3. Pivot table: Category vs Average Amount
print("\n3. Average order amount by category:")
print(ecommerce.groupby('product_category')['amount'].mean().sort_values(ascending=False))

# 4. Customer segments
customer_summary['segment'] = pd.cut(customer_summary['total_spent'],
                                     bins=[0, 500, 1000, float('inf')],
                                     labels=['Low', 'Medium', 'High'])

print("\n4. Customer Segments:")
print(customer_summary['segment'].value_counts())

# 5. Category preferences by segment
customer_orders = pd.merge(ecommerce, customer_summary[['customer_id', 'segment']], on='customer_id')
category_segment = pd.pivot_table(customer_orders,
                                  values='amount',
                                  index='segment',
                                  columns='product_category',
                                  aggfunc='count',
                                  fill_value=0)

print("\n5. Order count by Segment and Category:")
print(category_segment)
```
