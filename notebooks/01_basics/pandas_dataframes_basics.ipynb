## 1ï¸âƒ£ Creating DataFrames and Series

DataFrames are 2D labeled data structures (like tables).

```python
# Series (1D labeled array)
scores = pd.Series([85, 92, 78, 95, 88], 
                   index=['Alice', 'Bob', 'Charlie', 'David', 'Eve'])

print("Series:")
print(scores)
print(f"\nType: {type(scores)}")
print(f"Shape: {scores.shape}")
```

```python
# DataFrame from dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 28, 32],
    'Score': [85, 92, 78, 95, 88],
    'Pass': [True, True, True, True, True]
}

df = pd.DataFrame(data)
print("DataFrame:")
print(df)
```

```python
# DataFrame from NumPy array
np.random.seed(42)
array_data = np.random.rand(5, 3)

df_numpy = pd.DataFrame(array_data, 
                        columns=['Feature_1', 'Feature_2', 'Feature_3'],
                        index=['Sample_1', 'Sample_2', 'Sample_3', 'Sample_4', 'Sample_5'])

print("DataFrame from NumPy:")
print(df_numpy)
```

## 4ï¸âƒ£ Selecting Data

Access columns and rows in DataFrames.

```python
# Create sample dataset
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 28, 32],
    'Salary': [50000, 60000, 70000, 55000, 65000],
    'Department': ['Engineering', 'Marketing', 'Engineering', 'Sales', 'Engineering']
})

print("Full DataFrame:")
print(df)
print("\n" + "="*50 + "\n")

# Select single column (returns Series)
print("Names (Series):")
print(df['Name'])
print(f"Type: {type(df['Name'])}")

print("\n" + "="*50 + "\n")

# Select multiple columns (returns DataFrame)
print("Name and Age (DataFrame):")
print(df[['Name', 'Age']])
print(f"Type: {type(df[['Name', 'Age']])}")
```

```python
# loc: Label-based indexing
print("Row 0 using loc:")
print(df.loc[0])

print("\nRows 1-3 using loc:")
print(df.loc[1:3])

print("\nSpecific cells using loc:")
print(df.loc[1:3, ['Name', 'Salary']])
```

```python
# iloc: Position-based indexing
print("First row using iloc:")
print(df.iloc[0])

print("\nRows 1-3 using iloc:")
print(df.iloc[1:4])  # Note: exclusive end

print("\nFirst 3 rows, first 2 columns:")
print(df.iloc[:3, :2])
```

---

## 5ï¸âƒ£ Filtering Data

```python
# Boolean indexing
print("Employees with Age > 30:")
print(df[df['Age'] > 30])

print("\nEngineering Department:")
print(df[df['Department'] == 'Engineering'])

print("\nSalary >= 60000:")
print(df[df['Salary'] >= 60000])
```

```python
# Multiple conditions
print("Engineering AND Age < 30:")
print(df[(df['Department'] == 'Engineering') & (df['Age'] < 30)])

print("\nSalary > 60000 OR Age < 28:")
print(df[(df['Salary'] > 60000) | (df['Age'] < 28)])
```

```python
# isin() for multiple values
print("Engineering or Marketing:")
print(df[df['Department'].isin(['Engineering', 'Marketing'])])

# NOT in
print("\nNOT in Engineering:")
print(df[~df['Department'].isin(['Engineering'])])
```

---

## 6ï¸âƒ£ Adding and Modifying Columns

```python
# Add new column
df['Bonus'] = df['Salary'] * 0.1
print("Added Bonus column:")
print(df)

# Add column with condition
df['Senior'] = df['Age'] >= 30
print("\nAdded Senior column:")
print(df)

# Modify existing column
df['Salary'] = df['Salary'] * 1.05  # 5% raise
print("\nSalary after 5% raise:")
print(df[['Name', 'Salary']])
```

```python
# Drop columns
df_dropped = df.drop(['Bonus', 'Senior'], axis=1)
print("After dropping Bonus and Senior:")
print(df_dropped)
```

### ðŸŽ¯ ML Example: Feature Engineering

```python
# Create ML dataset
ml_df = pd.DataFrame({
    'age': [25, 30, 35, 40, 45, 50],
    'income': [30000, 45000, 55000, 70000, 85000, 95000],
    'credit_score': [650, 700, 720, 750, 780, 800]
})

print("Original features:")
print(ml_df)

# Feature engineering
ml_df['age_group'] = pd.cut(ml_df['age'], bins=[0, 30, 40, 100], 
                             labels=['young', 'middle', 'senior'])
ml_df['income_per_age'] = ml_df['income'] / ml_df['age']
ml_df['high_credit'] = ml_df['credit_score'] >= 750

print("\nWith engineered features:")
print(ml_df)
```

---

## 7ï¸âƒ£ Sorting and Ranking

```python
# Sort by column
print("Sorted by Salary (ascending):")
print(df.sort_values('Salary')[['Name', 'Salary']])

print("\nSorted by Salary (descending):")
print(df.sort_values('Salary', ascending=False)[['Name', 'Salary']])

# Sort by multiple columns
print("\nSorted by Department then Age:")
print(df.sort_values(['Department', 'Age'])[['Name', 'Department', 'Age']])
```

## ðŸŽ¯ Practice Exercises

### Exercise 1: Load and Explore Data

```python
# Create sample sales dataset
sales_data = pd.DataFrame({
    'product': ['A', 'B', 'C', 'D', 'E', 'A', 'B', 'C'],
    'quantity': [10, 15, 8, 20, 12, 18, 14, 9],
    'price': [100, 150, 80, 200, 120, 100, 150, 80],
    'region': ['East', 'West', 'East', 'North', 'South', 'West', 'East', 'North']
})

# Your tasks:
# 1. Display the first 3 rows
print("First 3 rows:")
print(sales_data.head(3))

# 2. Show shape and column names
print(f"\nShape: {sales_data.shape}")
print(f"Columns: {list(sales_data.columns)}")

# 3. Add 'revenue' column (quantity * price)
sales_data['revenue'] = sales_data['quantity'] * sales_data['price']
print("\nWith revenue column:")
print(sales_data)
```

### Exercise 2: Filter and Analyze

```python
# Filter high-revenue sales (revenue > 1500)
high_revenue = sales_data[sales_data['revenue'] > 1500]
print("High revenue sales (>1500):")
print(high_revenue)

# Find East region sales
east_sales = sales_data[sales_data['region'] == 'East']
print("\nEast region sales:")
print(east_sales)

# Top 3 by revenue
top3 = sales_data.nlargest(3, 'revenue')
print("\nTop 3 by revenue:")
print(top3)
```
